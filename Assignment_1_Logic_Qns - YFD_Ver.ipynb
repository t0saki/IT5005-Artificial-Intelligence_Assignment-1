### Q2b(i) Identify (with justification) two inference algorithms among tt_entails, resolution-refutation, forward chaining and backward chaining algorithms to solve this problem. [1 Marks]    



Write your answer for Q2b(i) here.

tt_entails：该算法系统地检查所有可能的模型（变量的真值赋值），以查看知识库是否包含查询。虽然它是完备且可靠的，但对于较大的问题，计算成本可能会很高。本题中变量较多，使用真值表蕴含算法的计算开销会比较大。

Resolution-refutation：这是一种通用的定理证明技术，可以处理任何类型的命题逻辑语句。它的工作原理是试图从查询的否定和知识库中推导出矛盾。适用于本题，因为题目中给出的条件和问题都可以转化为子句形式，且归结反驳算法能够有效地处理这些子句。
